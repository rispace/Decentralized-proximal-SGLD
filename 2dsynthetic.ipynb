Data Generation
$\delta_j \sim \mathcal{N}(0, 0.25),\quad \beta_j \sim \mathcal{N}(0, I),\quad y_j=\beta_*^{\top}X_j+\delta_j$

```python
def data_generator(
    num_data=1000, beta_star=None,
    sigma=0.5, seed=0
):
    """ 
    X: (num_data, dim), y: (num_data,)
    beta_star: (dim,)
    """
    beta_star = np.asarray(beta_star)
    dim = beta_star.size
    rng = np.random.default_rng(seed)
    X = rng.standard_normal(size=(num_data, dim))
    y = X @ beta_star + sigma * rng.standard_normal(size=num_data)
    return X, y
```

```python
beta_star = [0.5, 1.0]
X, y = data_generator(num_data=1000, beta_star=beta_star, sigma=0.5, seed=0)
```

```python
model = BayesianRegression(X=X, y=y)
```
